```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/IoT Network Intrusion Dataset.csv")
df
```

```python
df['Label'].value_counts()
```

Droping column with irrelevent data

```python
df1=df.drop(['Flow_ID','Src_IP','Src_Port','Dst_IP','Dst_Port','Timestamp','Cat','Sub_Cat'],axis=1)
```

```python
col=['Bwd_Pkts/b_Avg',
'Bwd_Byts/b_Avg',
'Fwd_Blk_Rate_Avg',
'Fwd_Pkts/b_Avg',
'Fwd_Byts/b_Avg',
'ECE_Flag_Cnt',
'CWE_Flag_Count',
'URG_Flag_Cnt',
'PSH_Flag_Cnt',
'RST_Flag_Cnt',
'SYN_Flag_Cnt',
'FIN_Flag_Cnt',
'Bwd_URG_Flags',
'Fwd_URG_Flags',
'Bwd_PSH_Flags',
'Fwd_PSH_Flags',
'FIN_Flag_Cnt',
'Fwd_Seg_Size_Min',
'Init_Fwd_Win_Byts',
'Bwd_Blk_Rate_Avg']

for i in col:
  print(i)
  print(df[i].value_counts())
  print('*'*100)
```

Droping column with only one unique value

```python
df1.drop(['Bwd_Pkts/b_Avg','Bwd_Byts/b_Avg','Fwd_Blk_Rate_Avg','Fwd_Pkts/b_Avg','Fwd_Byts/b_Avg','Fwd_URG_Flags','Fwd_PSH_Flags','Fwd_Seg_Size_Min',
          'Init_Fwd_Win_Byts','Bwd_Blk_Rate_Avg'],axis=1,inplace=True)
```

Droping columns having infin

```python
df1.drop(['Flow_Byts/s','Flow_Pkts/s'],axis=1,inplace=True)
```

```python
df1
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
y
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
y_resampled.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_resampled,y_resampled,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
sns.heatmap(df1.corr())
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

logistic= LogisticRegression()
logistic.fit(x_train, y_train)
y_pred = logistic.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(accuracy)
print(report)
```

Naive bayes

```python
from sklearn.naive_bayes import BernoulliNB

bernoulli = BernoulliNB()
bernoulli.fit(x_train, y_train)
y_pred = bernoulli.predict(x_test)

accuracy1 = accuracy_score(y_test, y_pred)
report1 = classification_report(y_test, y_pred)

print(accuracy1)
print(report1)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dec = DecisionTreeClassifier(criterion='gini')
dec.fit(x_train,y_train)
y_pred = dec.predict(x_test)

accuracy2 = accuracy_score(y_test, y_pred)
report2 = classification_report(y_test, y_pred)

print(accuracy2)
print(report2)
```

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.model_selection import cross_val_score

model = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

# Perform 5-fold cross-validation
# scores = cross_val_score(model, x, y, cv=5)

# print(scores)

accuracy3 = accuracy_score(y_test, y_pred)
report3 = classification_report(y_test, y_pred)

print(accuracy3)
print(report3)
```

gradient boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

gradient = GradientBoostingClassifier(random_state=10)

gradient.fit(x_train, y_train)
y_pred = gradient.predict(x_test)

accuracy4 = accuracy_score(y_test, y_pred)
report4 = classification_report(y_test, y_pred)

print(accuracy4)
print(report4)
```

Linear Discriminant Analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)

y_pred = lda.predict(x_test)

accuracy5 = accuracy_score(y_test, y_pred)
report5 = classification_report(y_test, y_pred)

print(accuracy5)
print(report5)
```

Quadratic Discriminant Analysis

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda = QuadraticDiscriminantAnalysis()
qda.fit(x_train,y_train)
y_pred = qda.predict(x_test)

accuracy6 = accuracy_score(y_test, y_pred)
report6 = classification_report(y_test, y_pred)

print(accuracy6)
print(report6)
```

```python
result=[accuracy,accuracy1,accuracy2,accuracy3,accuracy4,accuracy5,accuracy6]
name=['logistic','naive','decision','random','gradient','linear','quadratic']
```

```python
bars = plt.bar(name, result)
for bar, value in zip(bars, result):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(value,2),
             ha='center', va='bottom')

plt.show()
```

```python
a=df2.iloc[:,:-1]
b=df2.iloc[:,-1]
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
a_resampled, b_resampled = smote.fit_resample(a,b)
```

```python
b_resampled.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(a_resampled,b_resampled,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

Naive bayes

```python
from sklearn.naive_bayes import BernoulliNB

bernoulli = BernoulliNB()
bernoulli.fit(x_train, y_train)
y_pred = bernoulli.predict(x_test)

accuracy1 = accuracy_score(y_test, y_pred)
report1 = classification_report(y_test, y_pred)

print(accuracy1)
print(report1)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dec = DecisionTreeClassifier(criterion='gini')
dec.fit(x_train,y_train)
y_pred = dec.predict(x_test)

accuracy2 = accuracy_score(y_test, y_pred)
report2 = classification_report(y_test, y_pred)

print(accuracy2)
print(report2)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

accuracy3 = accuracy_score(y_test, y_pred)
report3 = classification_report(y_test, y_pred)

print(accuracy3)
print(report3)
```
